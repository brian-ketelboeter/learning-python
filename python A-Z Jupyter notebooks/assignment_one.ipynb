```python
from numpy.random import randn
```

```python
for k in range(1000, 10000, 100):
    counter = 0
    for n in range(1,k):
        x = randn()
        if -1 < x < 1:
            counter = counter + 1
    avg = counter / k
    print(k, avg)
```
